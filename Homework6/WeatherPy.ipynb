Once the latitudes and longitudes were generated, the citi names were found using the citipy api. Then I found the 
city weather using the openweather.org api.

Trend 1

We can observe that cities located around the 0 latitud -equator line- tend to be upper the zero temperature.

Trend 2

We can observe that most of cities tend to show humidity close to 100 no matter if located close to equator or no.

Trend 3

We can observe two group of Cloudiness: a group of cities with almost zero and a second group of cities between 80 and 100 %

I can conclude that because it is based on max temperature of the cities, the factor of local climate is determinant
for the results shown here.


```python
#### Dependencies
from citipy import citipy
import random
import pandas as pd
import json
import requests as req
import matplotlib.pyplot as plt

# Generate random coordinates for 600 cities as sample:
# latitudes: -90 under equator, 90 upper equator
# longitudes: -180 far west of meridian Zero, 
#             180 far east from meridian zero
coordinates = []
latitudes = []
longitudes = []

for i in range(1,901):
    lat = random.randint(-90,90)
    long = random.randint(-180, 180)
    latitudes.append(lat)
    longitudes.append(long)
    
coordinates = zip(latitudes,longitudes)

cities = []
for lat, lon in coordinates:
    cities.append(citipy.nearest_city(lat, lon))
citynames = []
for city in cities:
    citynames.append(city.city_name)

# make a set to remove duplicates
citynames = set(citynames)

```

```python
# Save Config information
url = 'http://api.openweathermap.org/data/2.5/weather'
api_key = 'a546c45f9bed5eff47e88c6a6052eb81'

weather_data = []
counter = 0
unit_type = 'imperial'
# Loop through the list of cities and perform a request for data on each
print('Begining Data Retrieval')
print('=========================')
for city in citynames:
    cityname = city.title()
    query_url = url + '?appid=' + api_key + '&q=' + cityname + '&units=' + unit_type
    # Get weather data
    # Do not include code 404 City not found 
    response = req.get(query_url).json()
    if response['cod'] != '404':
        weather_data.append(response)
        counter = counter + 1        
        print('Processing Record ' + str(counter) + ': ' +  response['name'])
        print(json.dumps(query_url, indent=4, sort_keys=True))
    
print('=========================')
print('Data Retrieval Complete')
print('=========================')
#print(weather_data)
```

```python
## Extract data needed from weather data list
country_list = []
city_list = []
lon_list = []
lat_list = []
temp_list = []
temp_max_list = []
hum_list = []
wind_list = []
cloud_list = []
date_list = []
for data in weather_data:
    city_list.append(data['name'])
    country_list.append(data['sys']['country'])
    lon_list.append(data['coord']['lon'])
    lat_list.append(data['coord']['lat'])
    temp_list.append(data['main']['temp'])
    temp_max_list.append(data['main']['temp_max'])
    hum_list.append(data['main']['humidity'])
    wind_list.append(data['wind']['speed'])
    cloud_list.append(data['clouds']['all'])
    date_list.append(data['dt'])
## 
##
# Create the city weather DataFrame 
city_weather_pd = pd.DataFrame()
city_weather_pd['country'] = country_list
city_weather_pd['city'] = city_list
city_weather_pd['longitud'] = lon_list
city_weather_pd['latitud'] = lat_list
city_weather_pd['temp_max'] = lon_list
city_weather_pd['humidity'] = hum_list
city_weather_pd['wind_speed'] = wind_list
city_weather_pd['cloudiness'] = cloud_list
city_weather_pd['date'] = date_list
city_weather_pd.head(10)
```

```python
# Build a scatter plot for Latitud vs Max Temperature data
fig, ax = plt.subplots(facecolor='white')
#plt.figure(figsize=(10, 6))
plt.scatter(city_weather_pd["latitud"], city_weather_pd["temp_max"], marker="o", edgecolor="black", facecolor="blue", linewidths=1)
plt.title("City Latitude Vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
ax.set_facecolor("whitesmoke")
plt.grid(True)
# Save the figure
plt.savefig("Latitud_MaxTemperature.png")
# Show plot
plt.show()
```

```python
# Build a scatter plot for Latitud vs Humidity data
fig, ax = plt.subplots(facecolor='white')
#plt.figure(figsize=(10, 6))
plt.scatter(city_weather_pd["latitud"], city_weather_pd["humidity"], marker="o", edgecolor="black", facecolor="blue", linewidths=1)
plt.title("City Latitude Vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
ax.set_facecolor("whitesmoke")
plt.grid(True)
# Save the figure
plt.savefig("Latitud_Humidity.png")
# Show plot
plt.show()
```

```python
# Build a scatter plot for Latitud vs Cloudiness data
fig, ax = plt.subplots(facecolor='white')
#plt.figure(figsize=(10, 6))
plt.scatter(city_weather_pd["latitud"], city_weather_pd["cloudiness"], marker="o", edgecolor="black", facecolor="blue", linewidths=1)
plt.title("City Latitude Vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
ax.set_facecolor("whitesmoke")
plt.grid(True)
# Save the figure
plt.savefig("Latitud_Cloudiness.png")
# Show plot
plt.show()
```

```python
# Build a scatter plot for Latitud vs Wind Speed data
fig, ax = plt.subplots(facecolor='white')
#plt.figure(figsize=(10, 6))
plt.scatter(city_weather_pd["latitud"], city_weather_pd["wind_speed"], marker="o", edgecolor="black", facecolor="blue", linewidths=1)
plt.title("City Latitude Vs. Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
ax.set_facecolor("whitesmoke")
plt.grid(True)
# Save the figure
plt.savefig("Latitud_WindSpeed.png")
# Show plot
plt.show()
```
